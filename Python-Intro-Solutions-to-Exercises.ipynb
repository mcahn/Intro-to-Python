## Exercise 18

- Write a function that converts Fahrenheit to Celsius
- Print a Fahrenheit to Celsius conversion table from 32 F to 212 F
- The forumla is:
  - C = 5/9 x (F - 32)
  
- Hints:
  - You will need a for loop using the range function.
  - Parentheses effect the order of arithmetic operations.

````python
def fahrenheitToCelsius(temp_in_f):
    temp_in_c = (5 / 9) * (temp_in_f - 32)
    return temp_in_c

for f in range(32, 213):
    print(f, fahrenheitToCelsius(f))
````

## Exercise 19

- Write a function to compute the number of violations per 1000 cars of a particular make
  - The function should take, as arguments, a make and a number of violations,
  - and return the number of violations per 1000 cars of that make.
- Print the number of violations for each make per 1000 vehicles
- Below is a dictionary of the number of cars on the road in Maryland for each make.

````python
carsOnRoad = {'ACURA': 27458,
'ALFA ROMEO': 3735,
'AUDI': 34839,
'BMW': 50956,
'BUICK': 35521,
'CADILLAC': 23297,
'CHEVROLET': 330381,
'CHRYSLER': 13857,
'DODGE': 43756,
'FIAT': 1339,
'FORD': 411035,
'GENESIS': 3585,
'GMC': 101759,
'HONDA': 266044,
'HYUNDAI': 139378,
'INFINITI': 16097,
'JAGUAR': 4320,
'JEEP': 176713,
'KIA': 125302,
'LAND ROVER': 14500,
'LEXUS': 50458,
'LINCOLN': 21280,
'MAZDA': 61199,
'MERCEDES': 69505,
'MINI': 5288,
'MITSUBISHI': 12197,
'NISSAN': 161317,
'PORSCHE': 12193,
'RAM': 127682,
'SMART': 15023,
'SUBARU': 136519,
'TESLA': 36800,
'TOYOTA': 347571,
'VOLKSWAGEN': 69933,
'VOLVO': 23770,}
````

````python
def normalizeViolations(make, violations):
    if make in carsOnRoad:
        return violations / (carsOnRoad[make] / 1000)
    else:
        return -1
    
    
openFile = open('Traffic_Violations-filtered.csv')

makeCounts = {}

for line in openFile:
    columns = line.split('|')
    
    make = columns[21]
    
    if make not in makeCounts:
        makeCounts[make] = 1
    else:
        makeCounts[make] = makeCounts[make] + 1
        
openFile.close()

for make in makeCounts:
    if make in carsOnRoad:
        norm = normalizeViolations(make, makeCounts[make])
        print(make, ': ', makeCounts[make], norm)
        
````

## Extra

### Print the makes in order of violations per 1000

- And format the output nicely

````python
# Put the make a violations per 1000 in a list of tuples.
# Each tuple will contain two items: the violations per 1000, and the make.
# The violations per 1000 is first, so the list can be easily sorted on that value.

violationsPer1000 = []

for make in makeCounts:
    if make in carsOnRoad:
        makeViolationsPer1000 = normalizeViolations(make, makeCounts[make])
        violationsPer1000.append((makeViolationsPer1000, make))
        
violationsPer1000.sort()

for vcount, make in violationsPer1000:
    print('{:20s}: {:7.3f}'.format(make, vcount))
````
